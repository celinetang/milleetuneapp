```python
import os
import glob
import pandas as pd

def create_dataframe(folder_path):
    image_paths = []
    folder_names = []
    supported_extensions = ('.jpg', '.jpeg', '.png', '.gif')  # Add or modify extensions as needed

    # Iterate through all files and subdirectories recursively
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(supported_extensions):
                image_paths.append(os.path.join(root, file))
                folder_names.append(os.path.basename(root))

    # Create a DataFrame from the extracted paths and folder names
    df = pd.DataFrame({'Image_Path': image_paths, 'Folder_Name': folder_names})
    return df

# Example usage:
folder_path = '/kaggle/input/insect-images-with-scientific-names/insects'
df = create_dataframe(folder_path)

# Print the DataFrame
df.head()
```

```python
d = pd.read_csv("/kaggle/input/insect-images-with-scientific-names/esoc_name_file.csv", encoding = 'unicode_escape')
merged_df = pd.merge(df, d, left_on='Folder_Name', right_on='scientific_name', how='left')
merged_df['final_class'] = merged_df['common_name'].apply(lambda x: str(x).split(" ")[-1])
merged_df['final_class'].value_counts()
```
